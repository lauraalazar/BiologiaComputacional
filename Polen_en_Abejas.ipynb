```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos ficticios
data = {
    'Familia': ['Urticaceae cecropia', 'Rhamnaceae', 'Dilleniaceae', 'Metastomataceae',
                'Cannabaceae Trema', 'Myrtaceae', 'Polypodiaceae', 'Malvaceae  Heliocarpus',
                'Fabaceae macrolobium', 'Poaceae', 'Malvaceae Bombacoideae', 'Loranthaceae oryctanthus'],
    'Conteo': [845, 120, 56, 39, 12, 21, 10, 7, 8, 6, 5, 4]
}

df = pd.DataFrame(data)
df['Porcentaje'] = df['Conteo'] / df['Conteo'].sum() * 100
df.sort_values(by='Porcentaje', ascending=False)
```

```python
plt.figure(figsize=(10,6))
plt.barh(df['Familia'], df['Porcentaje'], color='skyblue')
plt.xlabel('% de polen')
plt.title('Distribuci√≥n porcentual del polen por familia')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
